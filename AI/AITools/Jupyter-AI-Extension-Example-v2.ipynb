### Accessing Bedrock providers and interacting with corresponding LLMs without jupyter-ai extension

```python
# Set the HUIT Bedrock API key if not previously set above
if not "huit_bedrock_api_key" in globals():
    huit_bedrock_api_key = getpass.getpass("Enter your HUIT API key for AWS Bedrock: ")
```

```python
import requests

# Declare the configuration needed to access the provider and the model. 
# See https://portal.apis.huit.harvard.edu/docs/ais-bedrock-llm/1/overview#Getting-Started

BASE = "https://go.apis.huit.harvard.edu/ais-bedrock-llm/v2"
MODEL = "us.anthropic.claude-haiku-4-5-20251001-v1:0"

def ask_ai(prompt):
    url = f"{BASE}/model/{MODEL}/converse"
    headers = {
        "Content-Type": "application/json",
        "x-api-key": huit_bedrock_api_key,
    }
    payload = {
        "messages": [
            {"role": "user", "content": [{"text": prompt}]}
        ]
    }

    r = requests.post(url, headers=headers, json=payload, timeout=60)
    r.raise_for_status()
    data = r.json()

    # HUIT returns Bedrock Converse format
    return data["output"]["message"]["content"][0]["text"]

print(ask_ai("Write a Python function that adds two numbers"))
```
